```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer  #data stored in form of dictionary
```

```python
cancer.keys()
```

```python
print(cancer['target_names']) #Malignant - 0, Benign - 1
print(cancer['DESCR']) #description of data
#Total 32 Attributes
```

```python
df = pd.DataFrame(cancer['data'], columns= cancer['feature_names'])
df.head()
```

```python
#Difference between each value is very high. So to minimize used Standardcaler (Std = 1 and mean= 0) to sacle values
#MinMaxScaler(values between 0 to 1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
scaled_data
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
pca_data = pca.transform(scaled_data)
```

```python
print("Attributes reduced to 2 ",scaled_data.shape, pca_data.shape)
pca_data
```

```python
print(max(pca_data[:,0]),max(pca_data[:,1]))
print(min(pca_data[:,0]),min(pca_data[:,1]))
cancer['target']
```

```python
#Above 2 features as X and Y and accordingly plotted target
plt.figure(figsize=(8,6))  
plt.scatter(pca_data[:30,0],pca_data[:30,1],c=cancer['target'][:30]) #yellow - Benin, Purple - Malignant
plt.xlabel('First principle component')                              #Benign - 1, Malignant - 0
plt.ylabel('Second principle component')
```

#### https://chartio.com/learn/charts/what-is-a-scatter-plot/#:~:text=A%20scatter%20plot%20(aka%20scatter,to%20observe%20relationships%20between%20variables.

```python
plt.figure(figsize=(8,6))
plt.scatter(pca_data[:,0],pca_data[:,1],c=cancer['target'])
plt.xlabel('First principle component')                              
plt.ylabel('Second principle component')
```

```python
from sklearn.model_selection import train_test_split

train, test, train_labels, test_labels = train_test_split(pca_data, cancer['target'],
                                                         test_size = 0.30, random_state = 42)
```

```python
print(len(train_labels), len(test_labels))
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

gnb = GaussianNB()
naiveB_model = gnb.fit(train, train_labels)
```

```python
predictions = naiveB_model.predict(test)
print(predictions[:10], test_labels[:10])
```

```python
plt.figure(figsize=(35,7))
plt.plot(predictions, color='red')
plt.plot(test_labels)
plt.show()
```

```python
count = 0
for i in range(len(predictions)):
    if predictions[i] == test_labels[i]:
        count+=1
print(count)
print(count/len(predictions))
```

```python
print(accuracy_score(test_labels, predictions))
```
